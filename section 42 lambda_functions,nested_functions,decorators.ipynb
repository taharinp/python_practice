```python
def totalarea(l,b,h):
    def area(d1,d2):
        return d1*d2

    return 2*(area(l,b) +area(b,h)+area(l,h))



print(totalarea(10,2,3))
    
```

```python
def add(x,y):
    return x+y


def sub(x,y):
    return x-y

def airthmetic(f,x,y):
    return f(x,y)


sum=airthmetic(add ,10,3)
diff=airthmetic(sub ,10,3)
print(sum,diff)
```

closure functions

```python
def outer():
    msg='welcome'
    def Inner():
        print('+'*10)
        print(msg)
        print('+'*10)

    return Inner

f=outer() 
f()
```

decorator function

```python
def outer(f):
    def inner():
        print('+' *10)
        f()
        print('+' *10)
    return inner
@outer
def display():
    print('welcome')

#display=outer(display)
display()
    
```

```python
#or
def decorator(f):
    def inner():
        print('+' *10)
        f()
        print('+' *10)
    return inner
@decorator
def display():
    print('welcome')


display()
```
