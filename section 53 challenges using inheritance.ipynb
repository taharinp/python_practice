```python
import heapq

# Create empty heap
tasks = []

# Add tasks
heapq.heappush(tasks, (1, "High Priority Task"))
heapq.heappush(tasks, (3, "Low Priority Task"))
heapq.heappush(tasks, (2, "Medium Priority Task"))

print("Tasks Execution Order:")

while tasks:
    task = heapq.heappop(tasks)
    print(task)
```

```python
from datetime import datetime
import bisect

events = []   # sorted list of events

# Add event
def add_event():
    name = input("Enter event name: ")
    date_time = input("Enter date and time (DD-MM-YYYY HH:MM): ")

    dt = datetime.strptime(date_time,"%d-%m-%Y %H:%M")

    bisect.insort(events,(dt,name))

    print("Event Added Successfully")


# Show events
def show_events():

    if len(events)==0:
        print("No Events")
        return

    for e in events:
        print(e[0].strftime("%d-%m-%Y %H:%M"),"-",e[1])


# Menu
while True:

    print("\n1.Add Event")
    print("2.Show Events")
    print("3.Exit")

    choice=int(input("Enter choice: "))

    if choice==1:
        add_event()

    elif choice==2:
        show_events()

    elif choice==3:
        break

    else:
        print("Invalid Choice")
```

```python
from datetime import datetime

events = []

# Add Event
def add_event():
    name = input("Enter event name: ")
    date_str = input("Enter date and time (DD-MM-YYYY HH:MM): ")
    
    try:
        date_obj = datetime.strptime(date_str, "%d-%m-%Y %H:%M")
        events.append((name, date_obj))
        print("Event Added Successfully\n")
    except:
        print("Invalid Date Format\n")


# View Events
def view_events():
    if len(events) == 0:
        print("No events found\n")
        return
    
    print("\nAll Events:")
    for event in events:
        print(event[0], "-", event[1])


# Upcoming Events
def upcoming_events():
    now = datetime.now()
    
    print("\nUpcoming Events:")
    for event in events:
        if event[1] > now:
            print(event[0], "-", event[1])


# Menu
while True:
    print("\n1.Add Event")
    print("2.View Events")
    print("3.Upcoming Events")
    print("4.Exit")
    
    choice = input("Enter choice: ")
    
    if choice == "1":
        add_event()
        
    elif choice == "2":
        view_events()
        
    elif choice == "3":
        upcoming_events()
        
    elif choice == "4":
        break
        
    else:
        print("Invalid Choice")
```
