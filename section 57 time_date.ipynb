```python
from datetime import datetime, timedelta

today = datetime.today()

# Thursday = 3 (Monday=0)
days_back = (today.weekday() - 3) % 7

if days_back == 0:
    days_back = 7

last_thursday = today - timedelta(days=days_back)

print("Last Thursday was:", last_thursday.date())
```

```python
from datetime import datetime, timedelta

year = int(input("Enter year: "))

# Start from 1 January
date = datetime(year, 1, 1)

# Move to first Sunday
while date.weekday() != 6:   # Sunday = 6
    date += timedelta(days=1)

print("All Sundays:\n")

# Print all Sundays
while date.year == year:
    print(date.date())
    date += timedelta(days=7)
```
