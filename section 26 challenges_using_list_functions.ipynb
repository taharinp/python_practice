```python
import itertools as it
l1=['a','b','c','d']
perms=it.permutations(l1,r=2)
perm_list=list(perms)
print('permutations:',perm_list)
for t in perm_list:
    print(t)
```

```python
import statistics as st
```

```python
l1=[1,3,5,7,8,9,34,65,8,34,98,3,98,4,90]
mean=st.mean(l1)
median=st.median(l1)
mode=st.mode(l1)
print('mean:',mean)
print(median)
print(mode)
```
