```python
class Rational:

    def __init__(self, num, den):
        self.num = num
        self.den = den

    def __str__(self):
        return f"{self.num}/{self.den}"

    # Addition
    def __add__(self, other):
        n = self.num * other.den + self.den * other.num
        d = self.den * other.den
        return Rational(n, d)

    # Subtraction
    def __sub__(self, other):
        n = self.num * other.den - self.den * other.num
        d = self.den * other.den
        return Rational(n, d)

    # Multiplication
    def __mul__(self, other):
        n = self.num * other.num
        d = self.den * other.den
        return Rational(n, d)

    # Division
    def __truediv__(self, other):
        n = self.num * other.den
        d = self.den * other.num
        return Rational(n, d)


r1 = Rational(1, 2)
r2 = Rational(3, 4)

print("First =", r1)
print("Second =", r2)

print("Addition =", r1 + r2)
print("Subtraction =", r1 - r2)
print("Multiplication =", r1 * r2)
print("Division =", r1 / r2)
```

```python
inventory = {}

while True:
    print("\n1. Add Item")
    print("2. Remove Item")
    print("3. Update Quantity")
    print("4. Show Inventory")
    print("5. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        item = input("Enter item name: ")
        qty = int(input("Enter quantity: "))
        inventory[item] = qty
        print("Item added")

    elif choice == 2:
        item = input("Enter item name: ")
        if item in inventory:
            del inventory[item]
            print("Item removed")
        else:
            print("Item not found")

    elif choice == 3:
        item = input("Enter item name: ")
        qty = int(input("Enter new quantity: "))
        inventory[item] = qty
        print("Quantity updated")

    elif choice == 4:
        print("\nInventory List")
        for i in inventory:
            print(i, ":", inventory[i])

    elif choice == 5:
        break

    else:
        print("Invalid choice")
```

```python
class Payment:

    def pay(self, amount):
        print("Payment processing")


class CreditCard(Payment):

    def pay(self, amount):
        print("Paid", amount, "using Credit Card")


class UPI(Payment):

    def pay(self, amount):
        print("Paid", amount, "using UPI")


class Cash(Payment):

    def pay(self, amount):
        print("Paid", amount, "using Cash")


p1 = CreditCard()
p2 = UPI()
p3 = Cash()

p1.pay(1000)
p2.pay(500)
p3.pay(200)
```

```python
import pickle

# Student Data
student = {
    "name": "Ali",
    "roll": 101,
    "marks": 85
}

# Pickle (Save data)
file = open("student.dat", "wb")
pickle.dump(student, file)
file.close()

print("Data Saved")


# Unpickle (Read data)
file = open("student.dat", "rb")
data = pickle.load(file)
file.close()

print("Student Data:")
print(data)
```
