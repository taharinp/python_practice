URL PARSING

```python
url=input('enter url')
protocol=url[: url.find(':')]
dot1=url.find('.')
dot2=url.find('.',dot1+1)
domain=url[dot1+1:dot2]
page=url[url.find('/',dot2) :]
print(protocol)
print(domain)
print(page)
```

PALINDROME STRING

```python
s1='race car'
s2=s1.replace(' ','')
rev=s2[::-1]
if s2==rev:
    print('palindrome')
else:
    print('no palindrome')
```

ANAGRAMS

```python
S1='snooze alarms'
S2="alas,no more z's"
s1=S1.lower()
s2=S2.lower()
for x in s1:
    if x.isalpha():
        if s1.count(x)!=s2.count(x):
            print('not anagrams')
else:
    print('anagrams')
    
```

DATA CLEANING

```python

scan = 'These+notes#reveal9Newton seeking-out an{underlying structure to/the\pyramid}'
clean = ''

for x in scan:
    if x.isalpha() or x.isspace():
        clean = clean + x
    else:
        clean = clean + ' '
        
print(clean)
```
